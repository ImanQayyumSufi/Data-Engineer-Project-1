```python
def cleaning_data(df,columns):
    for column in columns:
        df[column] = df[column].str.replace(r'\bllc\b(?!\.)', 'LLC', regex=True, case=False)
        df[column] = df[column].str.replace(r'\binc\b(?!\.)', 'Inc.', regex=True, case=False)
        df[column] = df[column].str.replace(r'\bcorp\b(?!\.)', 'Corp.', regex=True, case=False)
        df[column] = df[column].str.replace(r'\bst\b(?!\.)', 'Street', regex=True, case=False)
        df[column] = df[column].str.replace('&Amp;', '&', regex=True, case=False)
        df[column] = df[column].str.replace(r'\b(?:ave|ave\.|Av)\b', 'Avenue', regex=True, case=False)
        df[column] = df[column].str.replace(r'\b(?:blvd|blvd\.)\b', 'Boulevard', regex=True, case=False)
        df[column] = df[column].str.replace(r'\bPLZ\b(?!\.)', 'Plaza', regex=True, case=False)
        df[column] = df[column].str.replace(r'\bRD\b(?!\.)', 'Road', regex=True, case=False)
        df[column] = df[column].str.replace(r'\bPL\b(?!\.)', 'Place', regex=True, case=False)
        df[column] = df[column].str.replace(r'\bTPKE\b(?!\.)', 'Turnpike', regex=True, case=False)
        df[column] = df[column].str.replace(r'\bPKWY\b(?!\.)', 'Parkway', regex=True, case=False)
        df[column] = df[column].str.replace(r'\bSQ\b(?!\.)', 'Square', regex=True, case=False)
        df[column] = df[column].str.replace(r'\bsLN\b(?!\.)', 'Lane', regex=True, case=False)
    
        # Replace numeric suffixes with lowercase
        df[column] = df[column].str.replace(r'(\d+)(st|nd|rd|th)', lambda x: f"{x.group(1)}{x.group(2).lower()}", regex=True, case=False)
    
        # Replace words with variations of apostrophe
        df[column] = df[column].replace(r"(\w)'S", r"\1's", regex=True)
        df[column] = df[column].replace(r"Don'T", "Don't", regex=True)

        # Remove { } from data
        df[column] = df[column].str.replace('[{}]', '', regex=True)

        # Remove unwanted _ inside a data
        df[column] = df[column].str.replace('_','', regex=True)
        
        # Replace all blank rows into "Not Available"
        df.fillna('Not Available', inplace=True)

        # Replace 'Na' and 'N/A' with 'Not Available' in the column
        df[column] = df[column].replace(['Na', 'N/A'], 'Not Available')
      
        return df
```

```python
def capitalize_data(df,columns):
    for column in columns:
        # Capitalize certain column
        df[column] = df[column].str.upper()
        return df
```

```python
def convert_datetime(df,columns):
    for column in columns:   
        #Change date into correct format
        df[column] = pd.to_datetime(df[column], errors='coerce')

        # Format with AM/PM and add :00 seconds
        df[column] = df[column].dt.strftime("%m/%d/%Y %I:%M:%S %p")

    return df

new_df = convert_datetime(new_df,['Time of Submission'])

#new_df
```

```python
# Function to clean the 'Building Number' column
def cleaning_building_number(value):
    try:
        # Try to parse as a date
        pd.to_datetime(value)
        return value
    except ValueError:
        try:
            # Try to parse as a float
            float_value = float(value)
            # Check if the float value has decimal places
            if float_value.is_integer():
                # If it's an integer, return the original value
                return value
            else:
                # If it has decimal places, return "Not Available"
                return "Not Available"
        except ValueError:
            # If not a valid date or float, replace with "Not Available"
            return "Not Available"

# Apply the clean_building_number function to the 'Building Number' column
new_df['Building Number'] = new_df['Building Number'].apply(cleaning_building_number)
#new_df
```

```python
def removing_column_with_dates(value):
    #Remove dates and put replace it as Not Available
    try:
        # Try to parse as a date
        pd.to_datetime(value, errors='raise')
        return "Not Available"
    except ValueError:
        # If not a valid date, replace with "Not Available"
        return value
```

```python
new_df['SLA Serial Number'] = new_df['SLA Serial Number'].apply(removing_column_with_dates)
```

```python
new_df['Building Number'] = new_df['Building Number'].apply(removing_column_with_dates)
new_df['Violation Code'] = new_df['Violation Code'].apply(removing_column_with_dates)
```

```python
import re

def cleaning_permit_column(df):
    # Define a list of words to exclude
    exclude_words = ['Restaurant', 'Spanish restaurant', 'RECORD #', 'CODE H25 CLASS/SUBCLASS FS', 'N/Abrewe', ', Code H25, Class fs'
                    '*Food processing license Agriculture and Markets','Grandfathered','General','Pizza','Nabrewer']

    # Use str.replace with regex for case-insensitive replacement
    pattern = '|'.join(map(re.escape, exclude_words))
    df['Food Service Establishment Permit'] = df['Food Service Establishment Permit'].str.replace(pattern, '', case=False, regex=True)

    # Make empty string to Not Available
    df['Food Service Establishment Permit'] = df['Food Service Establishment Permit'].replace('', 'Not Available')

    # Capitalize all data
    df['Food Service Establishment Permit'] = df['Food Service Establishment Permit'].str.upper()
   
    # Replace Na, N/A into Not Available  
    df['Food Service Establishment Permit'] = df['Food Service Establishment Permit'].replace(['Na', 'N/A', 'na','NA'], 'Not Available')
    return df

new_df = cleaning_permit_column(new_df)
```

```python
# Function to remove extra spaces in data
def remove_spaces(value):
    # Replace multiple spaces with a single space
    return ' '.join(value.split())
```

```python
def cleaning_inpspection_date(df,columns):
    for column in columns:
        # Replace '1/1/1900' with 'Not Available'
        df[column] = df[column].replace('1/1/1900', 'Not Available')

        # Convert the remaining dates to the desired format
        df[column] = pd.to_datetime(df[column], errors='coerce').dt.strftime('%d/%m/%Y')

        return df
```
